```python
# =========================================
# Search Module
# =========================================
class SearchModule:
    def __init__(self, keywords):
        # WHY: store the keyword entered by the user
        self.keywords = keywords

        # WHY: simulate source website (for demo purposes)
        self.sourceWebsite = "MockScholar"

    def searchSites(self):
        # WHY: simulate searching academic websites
        # In real system → this would call APIs or scrape websites

        print("Step 1: Searching for papers...")

        return [
            {"title": f"{self.keywords} in AI", "author": "Smith", "year": 2023},
            {"title": f"Advanced {self.keywords}", "author": "Ahmed", "year": 2022},
            {"title": f"{self.keywords} Applications", "author": "Lee", "year": 2024}
        ]


# =========================================
# Data Extractor
# =========================================
class DataExtractor:
    def __init__(self):
        # WHY: store raw and processed data (for future extension)
        self.rawHTML = None
        self.metadata = []

    def extractMetadata(self, data):
        # WHY: extract only useful information from raw data

        print("Step 2: Extracting important data...")

        extracted = []

        # Loop through each result
        for item in data:
            # Keep only needed fields
            extracted.append({
                "title": item["title"],
                "author": item["author"],
                "year": item["year"]
            })
        return extracted


# =========================================
# Processing Module
# =========================================
class ProcessingModule:
    def __init__(self):
        # WHY: store processed dataset
        self.dataset = []

    def filterData(self, data):
        # WHY: filter unwanted data (basic now, expandable later)

        print("Step 3: Filtering data...")

        return data

    def removeDuplicates(self, data):
        # WHY: remove duplicate research papers

        print("Step 4: Removing duplicates...")

        unique = {}

        # Use title as unique key
        for item in data:
            unique[item["title"]] = item

        return list(unique.values())

    def organizeResults(self, data):
        # WHY: rank and sort data (simulate intelligent decision)

        print("Step 5: Ranking results...")

        for item in data:
            # Calculate score:
            # Newer year = higher score
            # Longer title = slightly higher score
            score = (item["year"] - 2020) + len(item["title"]) / 10

            # Add score to each item
            item["score"] = round(score, 2)

        # Sort results from highest to lowest score
        data.sort(key=lambda x: x["score"], reverse=True)

        return data


# =========================================
# Storage System
# =========================================
import csv
class StorageSystem:
    def __init__(self):
        # WHY: define file format and path
        self.fileFormat = "CSV"
        self.storagePath = "results.csv"

    def saveCSV(self, data):
        # WHY: save results so user can open them in Excel

        print("Step 6: Saving results to CSV...")

        with open(self.storagePath, "w", newline="") as file:
            writer = csv.DictWriter(
                file,
                fieldnames=["title", "author", "year", "score"]
            )

            # Write column names
            writer.writeheader()
            # Write data rows
            writer.writerows(data)


# =========================================
# Agent (MAIN CONTROLLER)
# =========================================
class Agent:
    def __init__(self):
        # WHY: represent BDI model (conceptual)
        self.beliefs = "Knowledge about academic data"
        self.desires = "Find relevant research papers"
        self.intentions = "Execute workflow steps"

    def startSearch(self, keyword):
        # WHY: this function controls the full system workflow

        print("\n=== Intelligent Agent Started ===\n")

        # Create system components (modules)
        search_module = SearchModule(keyword)
        extractor = DataExtractor()
        processor = ProcessingModule()
        storage = StorageSystem()

        # Step-by-step execution
        raw_data = search_module.searchSites()
        extracted = extractor.extractMetadata(raw_data)
        filtered = processor.filterData(extracted)
        clean_data = processor.removeDuplicates(filtered)
        final_data = processor.organizeResults(clean_data)

        # Save results
        storage.saveCSV(final_data)

        print("\n=== Process Completed Successfully ===\n")

        return final_data


# =========================================
# RUN THE SYSTEM (MAIN PROGRAM)
# =========================================

print("Academic Intelligent Agent System")

# Ask user for keyword
keyword = input("Enter research topic: ")

# Create agent object
agent = Agent()

# Run the system
results = agent.startSearch(keyword)

# Print final results
print("Final Results:\n")
for r in results:
    print(r)
```
